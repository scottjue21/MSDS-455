### Data Cleaning

```python
# create year column and value column
df = df.melt(id_vars=['Country Name', 'Country Code', 'Series Name', 'Series Code'],
         var_name="Year", 
        value_name="Value")
```

```python
df.head()
```

```python
df['Country Code'].unique()
```

```python
# clean year field
df['Year']=df['Year'].str[:4].astype(int)
```

```python
df.info()
```

```python
# select last 10 years with data
df_10yr = df[(df['Year'] <= 2019) & (df['Year']>= 2009)]
```

```python
df_10yr.head()
```

```python
# subset renewable energy consumption by country
df_can_RE = df_10yr[(df_10yr['Country Code'] == 'CAN') & (df_10yr['Series Code'] == 'EG.FEC.RNEW.ZS')]
df_fra_RE = df_10yr[(df_10yr['Country Code'] == 'FRA') & (df_10yr['Series Code'] == 'EG.FEC.RNEW.ZS')]
df_deu_RE = df_10yr[(df_10yr['Country Code'] == 'DEU') & (df_10yr['Series Code'] == 'EG.FEC.RNEW.ZS')]
df_ita_RE = df_10yr[(df_10yr['Country Code'] == 'ITA') & (df_10yr['Series Code'] == 'EG.FEC.RNEW.ZS')]
df_jpn_RE = df_10yr[(df_10yr['Country Code'] == 'JPN') & (df_10yr['Series Code'] == 'EG.FEC.RNEW.ZS')]
df_gbr_RE = df_10yr[(df_10yr['Country Code'] == 'GBR') & (df_10yr['Series Code'] == 'EG.FEC.RNEW.ZS')]
df_usa_RE = df_10yr[(df_10yr['Country Code'] == 'USA') & (df_10yr['Series Code'] == 'EG.FEC.RNEW.ZS')]

# subset CO2 emissions by country
df_can_CE = df_10yr[(df_10yr['Country Code'] == 'CAN') & (df_10yr['Series Code'] == 'EN.ATM.CO2E.PC')]
df_fra_CE = df_10yr[(df_10yr['Country Code'] == 'FRA') & (df_10yr['Series Code'] == 'EN.ATM.CO2E.PC')]
df_deu_CE = df_10yr[(df_10yr['Country Code'] == 'DEU') & (df_10yr['Series Code'] == 'EN.ATM.CO2E.PC')]
df_ita_CE = df_10yr[(df_10yr['Country Code'] == 'ITA') & (df_10yr['Series Code'] == 'EN.ATM.CO2E.PC')]
df_jpn_CE = df_10yr[(df_10yr['Country Code'] == 'JPN') & (df_10yr['Series Code'] == 'EN.ATM.CO2E.PC')]
df_gbr_CE = df_10yr[(df_10yr['Country Code'] == 'GBR') & (df_10yr['Series Code'] == 'EN.ATM.CO2E.PC')]
df_usa_CE = df_10yr[(df_10yr['Country Code'] == 'USA') & (df_10yr['Series Code'] == 'EN.ATM.CO2E.PC')]
```

```python
# index data by country
df_can_RE['Pct_chg'] = df_can_RE['Value'].div(df_can_RE['Value'].iloc[0]).sub(1).mul(100)
df_fra_RE['Pct_chg'] = df_fra_RE['Value'].div(df_fra_RE['Value'].iloc[0]).sub(1).mul(100)
df_deu_RE['Pct_chg'] = df_deu_RE['Value'].div(df_deu_RE['Value'].iloc[0]).sub(1).mul(100)
df_ita_RE['Pct_chg'] = df_ita_RE['Value'].div(df_ita_RE['Value'].iloc[0]).sub(1).mul(100)
df_jpn_RE['Pct_chg'] = df_jpn_RE['Value'].div(df_jpn_RE['Value'].iloc[0]).sub(1).mul(100)
df_gbr_RE['Pct_chg'] = df_gbr_RE['Value'].div(df_gbr_RE['Value'].iloc[0]).sub(1).mul(100)
df_usa_RE['Pct_chg'] = df_usa_RE['Value'].div(df_usa_RE['Value'].iloc[0]).sub(1).mul(100)

df_can_CE['Pct_chg'] = df_can_CE['Value'].div(df_can_CE['Value'].iloc[0]).sub(1).mul(100)
df_fra_CE['Pct_chg'] = df_fra_CE['Value'].div(df_fra_CE['Value'].iloc[0]).sub(1).mul(100)
df_deu_CE['Pct_chg'] = df_deu_CE['Value'].div(df_deu_CE['Value'].iloc[0]).sub(1).mul(100)
df_ita_CE['Pct_chg'] = df_ita_CE['Value'].div(df_ita_CE['Value'].iloc[0]).sub(1).mul(100)
df_jpn_CE['Pct_chg'] = df_jpn_CE['Value'].div(df_jpn_CE['Value'].iloc[0]).sub(1).mul(100)
df_gbr_CE['Pct_chg'] = df_gbr_CE['Value'].div(df_gbr_CE['Value'].iloc[0]).sub(1).mul(100)
df_usa_CE['Pct_chg'] = df_usa_CE['Value'].div(df_usa_CE['Value'].iloc[0]).sub(1).mul(100)
```

```python
# merge dataframes
frames_RE = [df_can_RE, df_fra_RE, df_deu_RE, df_ita_RE, df_jpn_RE, df_gbr_RE, df_usa_RE]
frames_CE = [df_can_CE, df_fra_CE, df_deu_CE, df_ita_CE, df_jpn_CE, df_gbr_CE, df_usa_CE]

df_RE = pd.concat(frames_RE)
df_CE = pd.concat(frames_CE)
```

```python
# replace null values with 0
df_RE['Pct_chg'] = df_RE['Pct_chg'].fillna(0)
df_CE['Pct_chg'] = df_CE['Pct_chg'].fillna(0)
```

```python
# create list of G7 country names
countries = ['Canada', 'France', 'Germany', 'Italy', 'Japan', 'United Kingdom', 'United States']
```

### Create Line Chart to Show 10 year trend

```python
# plot data for renewable engergy consumption
fig,ax = plt.subplots()

for country in countries:
    ax.plot(df_RE[df_RE['Country Name']==country].Year,df_RE[df_RE['Country Name']==country].Pct_chg,label=country)

# label axes    
ax.set_xlabel("Year")
ax.set_ylabel("% Change in renewable energy consumption \n (% of total energy consumption)")
plt.xticks(np.arange(2009, 2020, 1))

# add legend
#ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

#label figure
plt.title('Renewable Energy Consumption by \n G7 Countries (2009-2019)')

#save image
plt.savefig('G7_renewable_energy_consumption.png', bbox_inches='tight')

plt.show()
```

```python
# plot data for CO2 Emissions
fig,ax = plt.subplots()

for country in countries:
    ax.plot(df_CE[df_CE['Country Name']==country].Year,df_CE[df_CE['Country Name']==country].Pct_chg,label=country)

# label axes    
ax.set_xlabel("Year")
ax.set_ylabel("% Change in C02 Emissions \n (metric tons per capita)")
plt.xticks(np.arange(2009, 2020, 1))

# add legend
ax.legend(loc='center left', bbox_to_anchor=(1.1, 0.25))


#label figure
plt.title('C02 Emissions by \n G7 Countries (2009-2019)')

# save image
plt.gcf()
plt.savefig('G7_CO2_emissions.png', bbox_inches='tight')

plt.show()
```
